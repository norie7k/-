```python
# ============= ä½ çš„æ¨¡å‹ä¸æ–‡ä»¶é…ç½®ï¼ˆæ”¹è¿™é‡Œï¼‰ =============
API_URL   = "https://ark.cn-beijing.volces.com/api/v3/chat/completions"
# API_KEY = "fcb27d69-1f94-46c7-b088-49f909e2f1d1"
# MODEL_ID  = "ep-20250630184125-mlgg8"     # ä½ çš„æ¨¡å‹ID
API_KEY = "51fb7806-f343-4767-8152-40a15f16838c" #è¾°å°‘å¯†é’¥
MODEL_ID = "ep-20250604162650-qfn2d"  #è¾°å°‘ ds

PROMPT_MD_PATH01 = Path("æç¤ºè¯1.md")      # æ¨¡å‹#1 system æç¤ºè¯ï¼ˆç­›ç›¸å…³ï¼‰
PROMPT_MD_PATH02 = Path("æç¤ºè¯2.md")      # æ¨¡å‹#2 system æç¤ºè¯ï¼ˆåšåˆ†ç±»ï¼‰

EXCEL_FILE       = Path("æ„å›¾åˆ†ç±»ç»“æœ_æ€»è¡¨test10.xlsx")   # è¾“å‡º Excel
BATCH_SIZE       = 100
SLEEP_BETWEEN    = 1   # æ¯æ‰¹ä¹‹é—´çš„é—´éš”ï¼Œé˜²æ­¢QPSè§¦å‘é™æµï¼›æŒ‰éœ€è°ƒæ•´
RETRIES          = 2
TEMPERATURE      = 0.25
MAX_TOKENS       = 32700
TIMEOUT_SEC      = 600
# =====================================================
```

## å¼€å§‹è¿è¡Œ

```python
from tqdm import tqdm  # âœ… æ–°å¢åº“

system_prompt01 = load_system_prompt(PROMPT_MD_PATH01)  # ç­›ç›¸å…³
system_prompt02 = load_system_prompt(PROMPT_MD_PATH02)  # åšåˆ†ç±»
create_intent_excel_styled(EXCEL_FILE)


total = len(jsonl_lines01)
if total == 0:
    print("æ²¡æœ‰å¯å¤„ç†çš„æ•°æ®ã€‚")
    # return

total_batches = (total + BATCH_SIZE - 1) // BATCH_SIZE
print(f"å‡†å¤‡å¤„ç† {total} æ¡ï¼Œå…± {total_batches} æ‰¹ï¼ˆæ¯æ‰¹ {BATCH_SIZE} æ¡ï¼‰ã€‚")

# âœ… ç”¨ tqdm åŒ…è£…å¾ªç¯
for b in tqdm(range(total_batches), desc="ğŸ”¥ æ‰¹å¤„ç†è¿›åº¦", unit="æ‰¹"):
    start = b * BATCH_SIZE
    end = min(start + BATCH_SIZE, total)
    batch_lines = jsonl_lines01[start:end]

    try:
            # --- æ¨¡å‹ #1ï¼šç­›ç›¸å…³ ---
        user_prompt1 = build_user_prompt_filter(batch_lines)
        output_filter = call_ark_chat_completions(
                api_url=API_URL,
                api_key=API_KEY,
                model=MODEL_ID,
                system_prompt=system_prompt01,
                user_prompt=user_prompt1,
                temperature=TEMPERATURE,
                max_tokens=MAX_TOKENS,
                timeout=TIMEOUT_SEC,
                retries=RETRIES,
            )
        print("ç¬¬ä¸€æ­¥å®Œæˆ")
        # print(output_filter)
        if not output_filter:
            tqdm.write(f"[æ‰¹æ¬¡ {b+1}] æ¨¡å‹#1 æ— æœ‰æ•ˆè¾“å‡ºï¼Œè·³è¿‡ã€‚")
            continue

            # --- æ¨¡å‹ #2ï¼šåˆ†ç±» ---
        user_prompt2 = build_user_prompt_classify(output_filter)
        # print(user_prompt2)
        output_classify = call_ark_chat_completions(
                api_url=API_URL,
                api_key=API_KEY,
                model=MODEL_ID,
                system_prompt=system_prompt02,
                user_prompt=user_prompt2,
                temperature=TEMPERATURE,
                max_tokens=MAX_TOKENS,
                timeout=TIMEOUT_SEC,
                retries=RETRIES,
            )
        # print(output_classify)
        # return

        # è½¬ DataFrame + å†™ Excel
        df_batch = jsonl_to_dataframe_with_intent(output_classify)
        append_df_to_excel_by_ca_styled(df_batch, EXCEL_FILE)  # â† æˆ‘ä¸Šä¸€æ¡ç»™ä½ çš„å†™å…¥å‡½æ•°

        tqdm.write(f"[æ‰¹æ¬¡ {b+1}] âœ… å†™å…¥ {len(df_batch)} æ¡ã€‚")

    except Exception as e:
        tqdm.write(f"[æ‰¹æ¬¡ {b+1}] âŒ å‡ºé”™ï¼š{e}")
        continue

    time.sleep(SLEEP_BETWEEN)

print("\nâœ… å…¨éƒ¨æ‰¹æ¬¡å¤„ç†å®Œæˆï¼")
```
